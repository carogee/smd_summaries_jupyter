```python
import numpy as np
import matplotlib.pyplot as plot
#import random
#import time
import h5py
import re
```

```python
filename = "/Users/cagee/Desktop/xcsl1024622_Run0336.h5"

exp = 'xcsl1024622'
run = 336

f = h5py.File(filename,'r')
collected_names = []


def print_name(name, obj):
    pattern = re.compile(r'ipm\d/sum')  # Regex pattern to match 'ipm' followed by any number of digits and then '/sum'
    #if re.search(pattern, name):
        #print(name)
    if pattern.search(name):
        collected_names.append(name)
       
with h5py.File(filename, 'r')  as h5f: # file will be closed when we exit from WITH scope
    h5f.visititems(print_name) # print all structure names

print("names",collected_names)


ipm_a = collected_names[0]
ipm_b = collected_names[1]

#print("ipmA", ipm_a)
#print("ipmB", ipm_b)
```
